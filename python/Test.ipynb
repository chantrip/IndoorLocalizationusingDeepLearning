```python
import itertools
import numpy;
import subprocess
import csv
import os.path

outputFileName = 'sae_test.csv'

def isExistedInCSV(field,value):
    if not os.path.isfile(outputFileName):
        return False
    with open(outputFileName) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row[field].replace(" ", "") == value.replace(" ", ""):
                return True
        return False

sae_size_space = [64,128,256,512]
for i in range(1,len(sae_size_space)+1):
    for j in list(itertools.product(sae_size_space,repeat=i)):
        sae_str = (",".join(str(i) for i in list(j)))
        if isExistedInCSV('sae_hidden_layers','['+sae_str+']'):
            print('Skipped :' + sae_str)
            continue
        print('Doing : ' + sae_str)
        %run sae_sweep.py -S $sae_str -C "64,128" -D 0.2 -N 8 --scaling 0.2 --batch_size 200
```
